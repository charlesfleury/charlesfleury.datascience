#Analyzing eBay Car Sales Data with Python

The goal of this project is to analyze 2016 data on used cars from eBay Kleinanzeigen, a section of the German eBay website. We'll be doing that using the Python libraries pandas and NumPy.

##Data cleaning

```python
import pandas as pd
import numpy as np

autos = pd.read_csv("autos.csv", encoding = "Latin-1")
```

```python
autos.describe(include="all")
```

We notice that the seller and offerType column almost have the same value for each row: they won't give us much information and we'll drop them.

Two of the interesting columns that we identified earlier (price  and odometer) are stored as strings, but we'd like them to be stored as numeric values instead. The next thing we'll do is convert them to numeric values.

```python
autos_price = autos["price"]
autos_price_clean = []
for p in autos_price:
    p = str(p)
    p = p.replace('$', '')
    p = p.replace(',', '')
    autos_price_clean.append(p)

autos["price"] = autos_price_clean   

autos["price"] = autos["price"].astype(np.float64)
    
autos_odometer = autos["odometer"]
autos_odometer_clean = []
for o in autos_odometer:
    o = str(o)
    o = o.replace("km", "")
    o= o.replace(',', '')
    autos_odometer_clean.append(o)
    
autos["odometer"] = autos_odometer_clean
    
autos["odometer"] = autos["odometer"].astype(np.int64)

autos = autos.rename(columns = {"odometer" :"odometer_km"})

autos.describe(include="all")
```

While we're at it, let's rename all the columns in snakecase format.

```python
autos = autos.rename(columns = {"dateCrawled" : "date_crawled", "offerType": "offer_type", "vehicleType": "vehicle_type", "yearOfRegistration": "year_of_registration", "powerPS": "power_ps", "monthOfRegistration": "month_of_registration", "fuelType": "fuel_type", "notRepairedDamage": "not_repaired_damage", "dateCreated": "date_created", "nrOfPictures": "nr_of_pictures", "postalCode": "postal_code", "lastSeen": "last_seen" })
```

##Data analysis and metrics

Now that we have a valid and workable dataset, let's find out some pertinent metrics and information.

First let's calculate an interesting value: the mean price per brand. We will do that for the top 25 brands by popularity in our set.

```python
#we create a dictionary which will store our mean prices per brand
top25_brands_mean_price = {}

#We select top 20 most popular car brands
top25_brands = autos["brand"].value_counts().head(25)
print("Most popular brands:")
print(top25_brands)

#iteration of the brands and calculation of the mean price
for b in top25_brands.index:
    brand_prices = autos.loc[autos["brand"]== b, "price"]
    mean_brand_price = brand_prices.mean()
    #adding the data to the dictionary
    top25_brands_mean_price[b] = mean_brand_price
    
print("Mean  price per brand:")
print(top25_brands_mean_price)
    
```

Here we see that the most popular brands are Volkswagen, BMW and Opel. We also printed the mean prices per brands, but we'll see better if we order them. Let's print this result in order of mean price:

```python
print(sorted(top25_brands_mean_price.items(),key = 
             lambda kv:(kv[1], kv[0])))
```

When observing this result, we can see that Renault, Fiat and Peugeot are the cheapest used car brands in this market. On the opposite side, Porsche, Sonstige Autos and Mini are the most expensive cars in our dataset among the top 25 brands by popularity. It's interesting to see that Volkswagen, the most popular brand in our set is located around the middle of the pack, while BMW and Mercedes-Benz, the second and third most popular brands, are the 6th and 5th most expensive brand in our top 25 list, respectively.

Another very interesting statistic to extract from our dataset for people looking for a bargain on a performance car is the horsepower/price ratio.

Let's create the list of the top 10 individual cars in the dataset based on this ratio.

```python
#calculate the ratio and add it as a new column
autos["ps_per_price"] = autos["power_ps"] / autos["price"]

#print information on the top 10 cars based on this ratio
top_10_ps_per_price = autos[["name", "brand", "model", "year_of_registration", "odometer_km", "price", "power_ps", "ps_per_price", "not_repaired_damage"]].sort_values(by = ["ps_per_price"], ascending=False).head(10)
top_10_ps_per_price
```

With this list, we can see that the car on which you'd pay the least per HP is a 1994 BMW Series 3 Compact. All of these cars are from the 90s or early 2000s and are on sale for under 1000$, which is a bargain if you're looking for performance. Notice that two of those cars have non-repaired damage.
